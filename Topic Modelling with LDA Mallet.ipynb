```python
#!python -m pip install -U gensim
#!python -m pip install -U nltk
#!pip install pyLDAvis==3.2.2

#nltk.download('punkt')

import gensim
import pandas as pd
from nltk.tokenize import word_tokenize
import nltk
#from gensim import corpora, models
import pyLDAvis.gensim
import os

df = pd.read_csv('cleaned_data.csv')
df.columns
```

```python
#clean_text is already tokensized text 
#docs=df['text'].apply(word_tokenize)

docs = df['clean_text'].apply(word_tokenize)
docs
```

```python
d= gensim.corpora.Dictionary(docs)

#convert docs to vec
v = [d.doc2bow(doc) for doc in docs]
print(v[0][0:20])
```

```python
os.environ.update({'MALLET_HOME':r'C:/mallet-2.0.8/'})

mallet_path = r'C:\mallet-2.0.8\bin\mallet'

ldamallet= gensim.models.ldamodel.LdaModel(corpus = v, 
                                           num_topics = 4, 
                                           id2word = d, 
                                           random_state = 100, 
                                           update_every = 1, 
                                           chunksize = 100, 
                                           passes = 10, 
                                           alpha = 'auto')

vecTop= ldamallet.show_topics()
for i in range(0, 8):
    print(vecTop[i])
```
